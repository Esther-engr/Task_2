```python
import pandas as pd
import seaborn as sns

dataset = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')
```

```python
dataset.info()

# This shows that the dataset contains 1000 rows with 8 columns.It also tells the Data types in each column and their various headings
```

```python
dataset.head(10)

#this data shows the test scores of a particular set of individuals, their genders, race, parents level of education, lunch type and preparation level
```

```python
dataset['race/ethnicity'].value_counts().plot.bar()

#this shows the number of individuals that took the test based on their race/ethnicity.
#clearly, group C and group A race has the highest and lowest number of test takers respectively
```

```python
dataset['parental level of education'].value_counts().sort_index().plot.bar()

#this shows the number of individuals that took the test based on their parents level of education.
#those whose parents has 'associate's degree' and 'some college', tops the chart
```

```python
dataset['race/ethnicity'].value_counts().sort_index().plot.line()

#this shows that group C and group A race has the highest and lowest number of test takers respectively
```

```python
dataset['math score'].value_counts().sort_index().plot.line()

#this shows that the math score of majority of the test takers is between 60-65
```

```python
dataset['math score'].value_counts().sort_index().plot.area()
```

```python
dataset['race/ethnicity'].value_counts().sort_index().plot.area()
```

```python
dataset['math score'].plot.hist()

#this shows the 'math scores' of all test takers
```

```python
dataset[dataset['math score'] > 20]['math score'].plot.hist()

#this shows the 'math scores' of all test takers greater than 20
```

```python
import seaborn as sns
```

```python
sns.distplot(dataset['math score'], bins=10, kde = False)

#this shows the 'math scores' of all the test takers
```

```python
sns.jointplot(x='math score', y='reading score', data=dataset[dataset['math score'] < 100])

# this is plotting two variables (math score & reading score) against one another
#this shows majority of the test takers scored between 70-80 in reading & 60-75 in math
```

```python
sns.jointplot(x='math score', y='reading score', data=dataset[dataset['math score'] > 20], kind = 'hex', gridsize = 20)

# this is plotting two variables (math score & reading score) against one another
#this shows majority of the test takers scored between 70-80 in reading & 60-75 in math
```

```python
dataset.head(5)
```

```python
df = dataset[dataset.lunch.isin(dataset.lunch.value_counts().head().index)]

sns.boxplot(
    x='lunch',
    y='math score',
    data=df
)
#this showed that those whose 'lunch' are 'standard' have higher 'math scores' compared to those whose 'lunch' are 'free/reduced'
```

```python
sns.violinplot(
    x='gender',
    y='math score',
    data=dataset[dataset.gender.isin(dataset.gender.value_counts()[:5].index)]
)

#this shows that the male gender has higher math scores than the female gender
```

```python
sns.countplot(dataset['race/ethnicity'])

#this shows that group C and group A race has the highest and lowest number of test takers respectively
```

```python
sns.jointplot(x='gender', y='math score', data=dataset)

#this shows that more female gender has lower math scores compared to the male gender
```

```python
sns.kdeplot(dataset['reading score'], dataset['writing score'])

# this shows that majority of the test takers scored between the range of 60-80 in both writing and reading
```
